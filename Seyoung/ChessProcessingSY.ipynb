```python
f= open('F:/spark_test/spark_exam/fide_board_data/onlymoves.pgn','r')
```

```python
lines = f.readlines()
Chessround=[]
for line in lines:
    roundmoves=[]
    if "TimeControl" in line:
        pass
    elif "Black" in line:
        pass
    elif "White" in line:
        pass
    else:
        line = line.strip()  # 줄 끝의 줄 바꿈 문자를 제거한다.
        linelist=line.split()
        for i in range(len(linelist)):
            if i % 7 in [1,4]:
                if i != len(linelist):
                    roundmoves.append(linelist[i])
        Chessround.append(roundmoves)
    
f.close()
```
