```python
f = open('./fide_board_data/moveagain.pgn', 'r')

lines = f.readlines()
Chessround=[]
for line in lines:
    roundmoves=[]
    line = line.strip()  # 줄 끝의 줄 바꿈 문자를 제거한다.
    linelist=line.split()
    for i in range(len(linelist)):
        if i % 7 in [1,4]:
            if i != len(linelist):
                 roundmoves.append(linelist[i])
    Chessround.append(roundmoves)
    
f.close()
```

```python
# 데이터 프레임 만들기
import pandas as pd
import numpy as np

df_chess = pd.DataFrame(Chessround)
df_chess
```

```python
# 10번째 수 까지 저장
# 처음 무브 10개만 저장
list1 = [x for x in range(0, 10)]

df_chess.columns
tot = df_chess[list1]
tot
```

```python
# 무브 별로 groupby 
final = tot.groupby([0], group_keys=True, squeeze=True, sort=True).size().sort_values(ascending=False)
final = pd.DataFrame(final)
final.head(10)
```
