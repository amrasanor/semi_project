```python
import pandas as pd
import numpy as np

chess = pd.read_csv("C:/chess-choco/semi_project/ChessData/fide_board_data/Chessdata.csv", index_col=None, header=None)
chess.head()
```

```python
# 백의 처음 수만 저장
list = [x for x in range(4)]

first4 = chess[list]
first4
```

```python
# groupby 후 정렬
f4_group = first4.groupby([0,1,2,3], group_keys=True, sort=True).size().sort_values(ascending=False)
f4_group = pd.DataFrame(f4_group)
f4_group
```

```python
# csv로 저장(리셋)
f4_group.to_csv('f4_group.csv')

# 다시 불러오면서 컬럼명 지정
f4_group_freq = pd.read_csv('./f4_group.csv')
f4_group_freq.columns = [0,1,2,3,'frequency']
f4_group_freq
```

```python
len(f4_group_freq)
```

```python
f4_group_freq.iloc[0][0]
```

```python
for i in range(len(f4_group_freq)):
    fm = f4_group_freq.iloc[i][0]
    
    if fm == 'e4':
        print("King's pawn opening")
    elif fm == 'd4':
        print("Queen's pawn opening")
    elif fm == 'Nf3':
        print("Reti opening")
    elif fm == 'c4':
        print("English opening")
    else :
        print("Irregular opening")
```
